```python
# Summary: Fraud Detection with Data Imbalance

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split, cross_val_score
import shap
import logging
import warnings
import os

# Suppress warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

# Setup Logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# Create a directory for saving visualizations
os.makedirs('visualizations', exist_ok=True)

# ========== Load + Analyze ==========
def load_and_analyze_dataset(filepath, target_column):
    """Load dataset and provide initial analysis"""
    try:
        data = pd.read_csv(filepath)
        logger.info(f"\nLoaded {filepath} with shape: {data.shape}")
        logger.info(f"Columns: {data.columns.tolist()}")
        
        # Check if target column exists
        if target_column not in data.columns:
            logger.error(f"Target column '{target_column}' not found in dataset")
            return None, None, None, None
        
        # Handle missing values
        missing_values = data.isnull().sum()
        if missing_values.sum() > 0:
            logger.warning(f"Missing values found:\n{missing_values[missing_values > 0]}")
            data = data.fillna(data.median(numeric_only=True))
            
        # Display class distribution
        class_counts = data[target_column].value_counts()
        total = len(data)
        class_ratio = (class_counts.min() / total) * 100

        logger.info(f"\nClass ratio for {filepath}: {class_ratio:.2f}%")
        logger.info(f"Total samples: {total}")
        logger.info(f"Class distribution:\n{class_counts}")

        # Check for non-numeric columns and convert if possible
        non_numeric = data.select_dtypes(exclude='number').columns.tolist()
        if non_numeric:
            logger.info(f"Non-numeric columns found: {non_numeric}")
            for col in non_numeric:
                if col != target_column:
                    try:
                        data[col] = pd.to_numeric(data[col], errors='coerce')
                    except:
                        logger.warning(f"Cannot convert column {col} to numeric, dropping it")
                        data = data.drop(columns=[col])
        
        X = data.drop(columns=[target_column])
        y = data[target_column]

        return X, y, data, class_counts
        
    except Exception as e:
        logger.error(f"Error loading dataset: {str(e)}")
        return None, None, None, None

# ========== Resample ==========
def resample_data(X, y):
    """Resample data using SMOTE and UnderSampling"""
    try:
        # Select only numeric features for resampling
        X_numeric = X.select_dtypes(include='number')
        
        # Apply SMOTE for oversampling
        over = SMOTE(random_state=42)
        X_over, y_over = over.fit_resample(X_numeric, y)
        over_counts = pd.Series(y_over).value_counts()
        logger.info(f"After oversampling: {over_counts}")

        # Apply RandomUnderSampler for undersampling
        under = RandomUnderSampler(random_state=42)
        X_under, y_under = under.fit_resample(X_numeric, y)
        under_counts = pd.Series(y_under).value_counts()
        logger.info(f"After undersampling: {under_counts}")

        return over_counts, under_counts, X_over, y_over, X_under, y_under
        
    except Exception as e:
        logger.error(f"Error during resampling: {str(e)}")
        # Return empty values as fallback
        empty_series = pd.Series(dtype='int64')
        return empty_series, empty_series, pd.DataFrame(), pd.Series(), pd.DataFrame(), pd.Series()

# ========== Visuals ==========
def plot_class_distribution(before, over, under, labels, title):
    """Plot class distributions before and after resampling"""
    try:
        fig, axs = plt.subplots(1, 4, figsize=(20, 5))
        fig.suptitle(title, fontsize=16)

        # Convert to list if pandas Series
        if isinstance(before, pd.Series):
            before = before.sort_index()
        if isinstance(over, pd.Series):
            over = over.sort_index()
        if isinstance(under, pd.Series):
            under = under.sort_index()

        # Plot original distribution
        axs[0].bar(range(len(before)), before.values, tick_label=before.index.astype(str), color='skyblue')
        axs[0].set_title('Original')
        axs[0].set_ylabel('Samples')
        
        # Plot oversampled distribution
        axs[1].bar(range(len(over)), over.values, tick_label=over.index.astype(str), color='lightgreen')
        axs[1].set_title('Oversampled')
        
        # Plot undersampled distribution
        axs[2].bar(range(len(under)), under.values, tick_label=under.index.astype(str), color='salmon')
        axs[2].set_title('Undersampled')
        
        # Plot pie chart
        axs[3].pie(before.values, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999'])
        axs[3].set_title('Class Ratio (Original)')
        
        plt.tight_layout()
        plt.savefig(f'visualizations/{title.replace(" ", "_")}_class_distribution.png')
        plt.show()
        
    except Exception as e:
        logger.error(f"Error plotting class distribution: {str(e)}")

def plot_corr_heatmap(df, title):
    """Plot correlation heatmap"""
    try:
        numeric = df.select_dtypes(include='number')
        if numeric.shape[1] < 2:
            logger.warning("Not enough numeric columns for correlation heatmap")
            return
            
        plt.figure(figsize=(12, 10))
        corr_matrix = numeric.corr()
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
        sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', annot=False, 
                    square=True, linewidths=.5, cbar_kws={"shrink": .5})
        plt.title(f'Correlation Heatmap - {title}')
        plt.tight_layout()
        plt.savefig(f'visualizations/{title.replace(" ", "_")}_correlation.png')
        plt.show()
        
    except Exception as e:
        logger.error(f"Error plotting correlation heatmap: {str(e)}")

def plot_feature_distributions(df, target_col, title):
    """Plot distribution of features by class"""
    try:
        numeric = df.select_dtypes(include='number').columns.tolist()
        if target_col in numeric:
            numeric.remove(target_col)
        
        if not numeric:
            logger.warning("No numeric features to plot distributions")
            return
            
        # Select at most 5 features to avoid too many plots
        for col in numeric[:5]:
            plt.figure(figsize=(10, 6))
            sns.kdeplot(data=df, x=col, hue=target_col, fill=True, common_norm=False)
            plt.title(f"{col} distribution by class - {title}")
            plt.savefig(f'visualizations/{title.replace(" ", "_")}_{col}_dist.png')
            plt.show()
            
    except Exception as e:
        logger.error(f"Error plotting feature distributions: {str(e)}")

def plot_boxplots(df, target_col, title):
    """Plot boxplots of features by class"""
    try:
        numeric = df.select_dtypes(include='number').columns.tolist()
        if target_col in numeric:
            numeric.remove(target_col)
            
        if not numeric:
            logger.warning("No numeric features to plot boxplots")
            return
            
        # Select at most 5 features to avoid too many plots
        for col in numeric[:5]:
            plt.figure(figsize=(10, 6))
            sns.boxplot(data=df, x=target_col, y=col)
            plt.title(f"{col} Boxplot by Class - {title}")
            plt.savefig(f'visualizations/{title.replace(" ", "_")}_{col}_boxplot.png')
            plt.show()
            
    except Exception as e:
        logger.error(f"Error plotting boxplots: {str(e)}")

def plot_pca(X, y, title):
    """Plot PCA visualization"""
    try:
        # Select only numeric columns
        X_numeric = X.select_dtypes(include='number')
        
        if X_numeric.shape[1] < 2:
            logger.warning("Not enough features for PCA")
            return
            
        # Scale data
        X_scaled = StandardScaler().fit_transform(X_numeric)
        
        # Apply PCA
        pca = PCA(n_components=min(2, X_numeric.shape[1]))
        X_pca = pca.fit_transform(X_scaled)
        
        # Create dataframe for visualization
        df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'] if X_numeric.shape[1] >= 2 else ['PC1'])
        df_pca['Target'] = y.values
        
        # Plot
        plt.figure(figsize=(10, 8))
        if X_numeric.shape[1] >= 2:
            sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Target', palette='Set1')
        else:
            sns.histplot(data=df_pca, x='PC1', hue='Target', palette='Set1')
            
        plt.title(f"PCA Projection - {title}")
        explained_var = pca.explained_variance_ratio_
        plt.xlabel(f"PC1 ({explained_var[0]:.2%} variance)")
        if X_numeric.shape[1] >= 2:
            plt.ylabel(f"PC2 ({explained_var[1]:.2%} variance)")
        plt.savefig(f'visualizations/{title.replace(" ", "_")}_pca.png')
        plt.show()
        
    except Exception as e:
        logger.error(f"Error plotting PCA: {str(e)}")

def plot_tsne(X, y, title):
    """Plot t-SNE visualization"""
    try:
        # Select only numeric columns
        X_numeric = X.select_dtypes(include='number')
        
        if X_numeric.shape[1] < 2:
            logger.warning("Not enough features for t-SNE")
            return
            
        # Limit to 1000 samples for speed if dataset is large
        if X_numeric.shape[0] > 1000:
            logger.info("Subsampling data for t-SNE to 1000 samples")
            X_subset, _, y_subset, _ = train_test_split(
                X_numeric, y, train_size=1000, random_state=42, stratify=y
            )
        else:
            X_subset, y_subset = X_numeric, y
            
        # Scale data
        X_scaled = StandardScaler().fit_transform(X_subset)
        
        # Apply t-SNE
        perplexity = min(30, X_subset.shape[0] // 5)  # Adjust perplexity based on sample size
        tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity)
        X_tsne = tsne.fit_transform(X_scaled)
        
        # Create dataframe for visualization
        df_tsne = pd.DataFrame(X_tsne, columns=['Dim1', 'Dim2'])
        df_tsne['Target'] = y_subset.values
        
        # Plot
        plt.figure(figsize=(10, 8))
        sns.scatterplot(data=df_tsne, x='Dim1', y='Dim2', hue='Target', palette='Set2')
        plt.title(f"t-SNE Visualization - {title}")
        plt.savefig(f'visualizations/{title.replace(" ", "_")}_tsne.png')
        plt.show()
        
    except Exception as e:
        logger.error(f"Error plotting t-SNE: {str(e)}")

def feature_importance(X, y, model_name, title):
    """Plot feature importance"""
    try:
        # Select only numeric columns
        X_numeric = X.select_dtypes(include='number')
        
        if X_numeric.shape[1] < 1:
            logger.warning("No numeric features for importance analysis")
            return
            
        # Initialize model
        if model_name == 'RF':
            model = RandomForestClassifier(n_estimators=100, random_state=42)
        else:  # XGB
            model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
            
        # Fit model
        model.fit(X_numeric, y)
        
        # Get feature importances
        importances = pd.Series(model.feature_importances_, index=X_numeric.columns).sort_values(ascending=False)
        
        # Plot top features
        n_features = min(10, len(importances))
        plt.figure(figsize=(12, 8))
        sns.barplot(x=importances.values[:n_features], y=importances.index[:n_features], palette="viridis")
        plt.title(f"Top {n_features} Feature Importance - {model_name} - {title}")
        plt.xlabel("Importance")
        plt.tight_layout()
        plt.savefig(f'visualizations/{title.replace(" ", "_")}_{model_name}_importance.png')
        plt.show()
        
        # SHAP Analysis for better interpretability
        # Limit to 100 samples for SHAP to improve speed
        if X_numeric.shape[0] > 100:
            X_sample, _, y_sample, _ = train_test_split(
                X_numeric, y, train_size=100, random_state=42, stratify=y
            )
        else:
            X_sample = X_numeric
        
        # Create SHAP explainer
        try:
            explainer = shap.TreeExplainer(model)
            shap_values = explainer.shap_values(X_sample)
            
            # For XGBoost, shap_values is a list for binary classification
            if isinstance(shap_values, list):
                shap_values = shap_values[1]  # Take positive class values
                
            plt.figure(figsize=(12, 8))
            shap.summary_plot(shap_values, X_sample, plot_type="bar", show=False)
            plt.title(f"SHAP Feature Importance - {model_name} - {title}")
            plt.tight_layout()
            plt.savefig(f'visualizations/{title.replace(" ", "_")}_{model_name}_shap.png')
            plt.show()
        except Exception as e:
            logger.error(f"Error in SHAP analysis: {str(e)}")
        
        return importances
        
    except Exception as e:
        logger.error(f"Error analyzing feature importance: {str(e)}")
        return pd.Series()

def classification_visuals(X, y, title):
    """Generate classification model evaluation visualizations"""
    try:
        # Select only numeric columns
        X_numeric = X.select_dtypes(include='number')
        
        if X_numeric.shape[1] < 1:
            logger.warning("No numeric features for classification")
            return
            
        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X_numeric, y, test_size=0.3, random_state=42, stratify=y
        )
        
        # Train model
        model = RandomForestClassifier(n_estimators=100, random_state=42)
        model.fit(X_train, y_train)
        
        # Predictions
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:,1] if len(model.classes_) == 2 else None
        
        # Cross-validation score
        cv_score = cross_val_score(model, X_numeric, y, cv=5, scoring='accuracy')
        logger.info(f"Cross-validation scores: {cv_score}")
        logger.info(f"Mean cross-validation score: {cv_score.mean():.4f}")
        
        # Confusion Matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f"Confusion Matrix - {title}")
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.savefig(f'visualizations/{title.replace(" ", "_")}_confusion_matrix.png')
        plt.show()
        
        # Classification Report
        logger.info(f"\nClassification Report for {title}:")
        report = classification_report(y_test, y_pred)
        print(report)
        
        # For binary classification, plot ROC and PR curves
        if y_proba is not None:
            # ROC Curve
            fpr, tpr, _ = roc_curve(y_test, y_proba)
            roc_auc = auc(fpr, tpr)
            plt.figure(figsize=(8, 6))
            plt.plot(fpr, tpr, label=f"ROC curve (area = {roc_auc:.2f})")
            plt.plot([0, 1], [0, 1], linestyle='--')
            plt.title(f"ROC Curve - {title}")
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.legend()
            plt.savefig(f'visualizations/{title.replace(" ", "_")}_roc_curve.png')
            plt.show()
            
            # Precision-Recall Curve
            precision, recall, _ = precision_recall_curve(y_test, y_proba)
            plt.figure(figsize=(8, 6))
            plt.plot(recall, precision, label='Precision-Recall curve')
            plt.title(f"Precision-Recall Curve - {title}")
            plt.xlabel('Recall')
            plt.ylabel('Precision')
            plt.legend()
            plt.savefig(f'visualizations/{title.replace(" ", "_")}_pr_curve.png')
            plt.show()
            
    except Exception as e:
        logger.error(f"Error generating classification visuals: {str(e)}")

# ========== Summary Generator ==========
def generate_conclusions(dataset_name, class_counts, over_counts, important_features):
    """Generate conclusions based on analysis"""
    try:
        logger.info(f"\n Conclusion for {dataset_name}:")
        
        # Calculate class imbalance
        ratio = class_counts.min() / class_counts.sum() * 100
        imbalance_note = "HIGHLY imbalanced" if ratio < 10 else "MODERATELY balanced"
        logger.info(f"- The dataset is {imbalance_note} with a minority class ratio of {ratio:.2f}%.")
        
        # Resampling effect
        logger.info(f"- After oversampling, both classes are balanced with {over_counts.min()} samples each.")
        
        # Feature importance
        if len(important_features) > 0:
            top_features = important_features[:min(5, len(important_features))].index.tolist()
            logger.info(f"- Top influential features include: {', '.join(top_features)}")
        else:
            logger.info("- No feature importance available.")
            
        # Visualization insights
        logger.info("- PCA and t-SNE plots indicate" + 
                (" good" if ratio > 10 else " weak") + " class separation.")
                
    except Exception as e:
        logger.error(f"Error generating conclusions: {str(e)}")

# ========== MAIN RUNNER ==========
def run_all(filepath, target, labels, dataset_name):
    """Run complete analysis pipeline on dataset"""
    try:
        logger.info(f"\n{'='*50}\nProcessing {dataset_name}\n{'='*50}")
        
        # Load and analyze dataset
        X, y, df, class_counts = load_and_analyze_dataset(filepath, target)
        if X is None:
            logger.error(f"Could not load dataset {filepath}")
            return
            
        # Resample data
        over_counts, under_counts, X_over, y_over, X_under, y_under = resample_data(
            X, y
        )
        
        # Generate visualizations
        plot_class_distribution(class_counts, over_counts, under_counts, labels, dataset_name)
        plot_corr_heatmap(df, dataset_name)
        plot_feature_distributions(df, target, dataset_name)
        plot_boxplots(df, target, dataset_name)
        plot_pca(X, y, dataset_name)
        plot_tsne(X, y, dataset_name)
        
        # Get feature importance
        rf_importance = feature_importance(X, y, 'RF', dataset_name)
        xgb_importance = feature_importance(X, y, 'XGB', dataset_name)
        
        # Model evaluation
        classification_visuals(X, y, dataset_name)
        
        # Generate conclusions
        generate_conclusions(dataset_name, class_counts, over_counts, 
                           rf_importance if not rf_importance.empty else xgb_importance)
                           
        logger.info(f"Completed analysis for {dataset_name}")
        
    except Exception as e:
        logger.error(f"Error in pipeline for {dataset_name}: {str(e)}")

# ========== Main Execution ==========
if __name__ == "__main__":
    try:
        # Check if files exist before running
        files_to_check = ["creditcard.csv", "twitter_human_bots_dataset.csv"]
        missing_files = [f for f in files_to_check if not os.path.exists(f)]
        
        if missing_files:
            logger.error(f"Missing files: {missing_files}")
            logger.info("Please download the required datasets first.")
        else:
            # Run analysis on both datasets
            run_all("creditcard.csv", "Class", ["Not Fraud", "Fraud"], "Credit Card Fraud")
            run_all("twitter_human_bots_dataset.csv", "account_type", ["Human", "Bot"], "Twitter Bot Detection")
    except Exception as e:
        logger.error(f"Error in main execution: {str(e)}")
```
